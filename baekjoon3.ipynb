```java
//10870 피보나치 수열
import java.util.Scanner;

public class Main {

	static int result(int n) {
		if(n==1) {
			return 1;
		}
		else if(n==0) {
			return 0;
		}
		else{
			return result(n-1)+result(n-2);
		}


	}
	public static void main(String[] args)  {
		Scanner sc=new Scanner(System.in);
		int n=sc.nextInt();
		
		int fina=result(n);
		System.out.print(fina);
		
	}
}
```

### 10 브루트 포스

```java
//2798
//3중 반복문을 사용하면 시간 초과가 뜰 것 같았는데 그렇지는 않았다 하지만 효율적으로 5 컴비네이션 3을 계산하는 방법 연구 필요
import java.util.Scanner;
import java.lang.Math;

public class Main{

	public static void main(String[] args) {
		Scanner sc=new Scanner(System.in);
		int n=sc.nextInt();
		int m=sc.nextInt();
		int[] arr=new int[n];
		for(int i=0;i<n;i++) {
			arr[i]=sc.nextInt();
		}
		int max=0;
		int res=m;
		int re=0;
		for(int i=0;i<n;i++) {
			for(int j=0;j<n;j++) {
				for(int z=0;z<n;z++) {
					if(i!=j&&j!=z&&i!=z) {
						re=arr[i]+arr[j]+arr[z];
						if((Math.abs(re-m)<res)&(re<=m)) {
							res=Math.abs(re-m);
							max=re;
						}
					}
					
					
			
				}
			}
		}
		System.out.print(max);
	}
	

}
```

```java
//2231
//저번주 스터디 때 분해합을 구할 때 값보다 작은 수를 모두 대입하는 브루트 포스 방법을 사용한 문제 풀이를 접해서 쉽게 해결 가능했다
import java.util.Scanner;
import java.lang.Math;

public class Main{
	static int result(int n) {
		int fina=n;
		while(n>0) {
			fina+=n%10;
			n=n/10;
		}
		
		return fina;
	}
	public static void main(String[] args) {
		Scanner sc=new Scanner(System.in);
		int n=sc.nextInt();
		
		int min=n;
		for(int i=1;i<=n;i++) {
			if(n==result(i)) {
				if(min>i)
					min=i;
			}
		}
		if(min==n) {
			System.out.println(0);
		}
		else {
			System.out.println(min);
		}

		
	}
	

}

```

### 11 정렬

```java
//2750 
//자바에서 버블 소트를 처음 구현해 보았는데 포인터 대신, 인덱스를 swap함수에 사용하여 구현한다는 점이 새로웠다
import java.util.Scanner;

public class Main{
	static void swap(int[] a,int i,int j) {
		int temp=a[i];
		a[i]=a[j];
		a[j]=temp;
	}
	static void result(int a[], int n) {
		for(int i=0;i<n-1;i++) {
			for(int j=0;j<n-1;j++) {
				if(a[j]>a[j+1])
					swap(a,j,j+1);
			}
		}
		for(int i=0;i<n;i++) {
			System.out.println(a[i]);
		}
	}
	public static void main(String[] args) {
		Scanner sc=new Scanner(System.in);
		int n=sc.nextInt();
		
		int[] arr=new int[n];
		for(int i=0;i<n;i++) {
			arr[i]=sc.nextInt();
		}
		result(arr,n);
		
	}
	

}

```

```java
//25305
//앞에서 구현한 버블소트를 활용하여 해결한 문제
import java.util.Scanner;

public class Main{
	static void swap(int[] a,int i,int j) {
		int temp=a[i];
		a[i]=a[j];
		a[j]=temp;
	}
	static void result(int a[], int n,int m) {
		for(int i=0;i<n-1;i++) {
			for(int j=0;j<n-1;j++) {
				if(a[j]>a[j+1])
					swap(a,j,j+1);
			}
		}
		
		System.out.println(a[n-m]);
	}
	public static void main(String[] args) {
		Scanner sc=new Scanner(System.in);
		int n=sc.nextInt();
		int m=sc.nextInt();
		
		int[] arr=new int[n];
		for(int i=0;i<n;i++) {
			arr[i]=sc.nextInt();
		}
		result(arr,n,m);
		
	}
	

}
```

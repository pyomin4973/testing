```java
//스택 10828번
import java.io.*;
import java.util.Scanner;

public class Main {
	static int[] stack;
	static int index;
	
	static void push(int x) {
		index++;
		stack[index]=x;
	}
	static int pop() {
		if(index==-1)
			return -1;
		else {
			return stack[index--];
		}
	}
	static int size() {
		if(index==-1)
			return 0;
		else
			return index+1;
	}
	static boolean empty() {
		if(index==-1)
			return false;
		else
			return true;

	}
	static int top() {
		if(index==-1)
			return -1;
		else
			return stack[index];
	}
    public static void main(String[] args) {
        Scanner sc=new Scanner(System.in);
        StringBuilder sb=new StringBuilder();
        
        int in=sc.nextInt();
        stack=new int[in];
        index=-1;
        
        for(int i=0;i<in;i++) {
        	String one=sc.next();
        	if(one.equals("top")) {
        		int check=top();
        		sb.append(check+"\n");
        		//System.out.println(check);
        	}
        	else if(one.equals("pop")) {
        		int check=pop();
        		//System.out.println(check);
        		sb.append(check+"\n");
        	}
        	else if(one.equals("size")) {
        		int check=size();
        		sb.append(check+"\n");
        	}
        	else if(one.equals("empty")) {
        		boolean check=empty();
        		if(check==false)
        			sb.append(1+"\n");
        		else {
        			sb.append(0+"\n");
        		}
        	}
        	else {//push
        		String two=sc.next();
        		int fin=Integer.parseInt(two);
        		push(fin);
        	}
        }
        System.out.println(sb);
    }
    
    
}
```

```java
//큐 10845번
import java.io.*;
import java.util.Scanner;

public class Main {
	static int[] queue;
	static int head;
	static int tail;
	
	static void push(int x) {
		tail++;
		queue[tail]=x;
	}
	static int pop() {
		if(head>tail)
			return -1;
		else {
			return queue[head++];
		}
	}
	static int size() {
		return tail-head+1;
	}
	static int empty() {
		if(head>tail)
			return 1;
		else
			return 0;

	}
	static int front() {
		if(head>tail)
			return -1;
		else
			return queue[head];
	}
	static int back() {
		if(head>tail)
			return -1;
		else
			return queue[tail];
	}
    public static void main(String[] args) {
        Scanner sc=new Scanner(System.in);
        StringBuilder sb=new StringBuilder();
        
        int in=sc.nextInt();
        queue=new int[in];
        tail=-1;
        head=0;
        
        for(int i=0;i<in;i++) {
        	String one=sc.next();
        	if(one.equals("front")) {
        		int check=front();
        		sb.append(check+"\n");
        		//System.out.println(check);
        	}
        	else if(one.equals("back")) {
        		int check=back();
        		sb.append(check+"\n");
        		//System.out.println(check);
        	}
        	else if(one.equals("pop")) {
        		int check=pop();
        		//System.out.println(check);
        		sb.append(check+"\n");
        	}
        	else if(one.equals("size")) {
        		int check=size();
        		sb.append(check+"\n");
        	}
        	else if(one.equals("empty")) {
        		int check=empty();
        		sb.append(check+"\n");
        	}
        	else {//push
        		String two=sc.next();
        		int fin=Integer.parseInt(two);
        		push(fin);
        	}
        }
        System.out.println(sb);
    }
    
    
}
```

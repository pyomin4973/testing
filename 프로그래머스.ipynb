```python
# 가장 가까운 같은 글자
def solution(s):
  answer=[]
  for i in range(len(s)):
    check=s[:i]
    check=''.join(check[::-1]) # 문자열 뒤집기
    index=check.find(s[i])+1 # 문자열 중 char 위치 찾기
    if index==0:
      index=-1
    answer.append(index)

  return answer
```

```python
# 키패드 누르기(거리 구하기, xy축)
def getLength(hand, i):
    keypad = {1:[0,0], 2:[1,0], 3:[2,0],
              4:[0,1], 5:[1,1], 6:[2,1],
              7:[0,2], 8:[1,2], 9:[2,2],
            '*':[0,3], 0:[1,3], '#':[2,3]}
    
    x_hand, y_hand = keypad[hand]
    x_num, y_num = keypad[i]
    
    return abs(x_hand-x_num)+abs(y_hand-y_num)
```

```python
# BFS, 만들 수 있는 수 모두 구한 후, 조건에 부합하면 +1
def solution(numbers, target):
    answer = 0
    check=[0]
    for i in numbers:
        tmp=[]
        for j in check:
            tmp.append(j+i)
            tmp.append(j-i)
        check=tmp
    for i in check:
        if i ==target:
            answer+=1
    return answer
```

```python
# 게임 맵 최단거리 구하기(queue 사용, DFS)
from collections import deque
def solution(maps):
    answer = 0
    dx,dy=[-1,1,0,0],[0,0,-1,1]
    q=deque([(0,0)])
    
    visit=[[0]*len(maps[0]) for _ in range(len(maps))]
    visit[0][0]=1
    maps[0][0]=0
    
    while q:
        x,y=q.popleft()
        for k in range(4):
            nx,ny=x+dx[k], y+dy[k]
            if 0<=nx<len(maps) and 0<=ny<len(maps[0]):
                if maps[nx][ny]==1 and not visit[nx][ny]:
                    visit[nx][ny]=visit[x][y]+1
                    q.append((nx,ny))
    answer=-1
    if visit[-1][-1]:
        return visit[-1][-1]
    return answer
```
